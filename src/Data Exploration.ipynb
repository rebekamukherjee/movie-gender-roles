```python
with open('../data/processed/movies.pkl', 'rb') as movies_pkl:
    movies = pickle.load(movies_pkl)
```

```python
def returnGenre(s):
    p = re.finditer(r'name', s)
    position = [m.start(0) for m in p]
    allGenres = set()
    for pos in position:
        genre = ''
        for i in range(pos+8, len(s)):
            if s[i] == '\'':
                break
            genre += s[i]
        allGenres.add(genre)
    return allGenres
```

# 1. Find number of movies where number of female cast members was more than number of male cast members

```python
en_movie_count = 0
en_movie_count_female = 0
female_genres = set()
for movie in movies:
    if movie.lang == 'en':
        en_movie_count += 1
        num_female = len(movie.cast.get_female_cast())
        num_male = len(movie.cast.get_male_cast())
        if num_female > num_male:
            en_movie_count_female += 1
            female_genres.update(returnGenre(movie.genres))
print en_movie_count_female, en_movie_count
print en_movie_count_female*100.0/en_movie_count
```

# 2. Find female-to-male cast ratio for each genre

```python
uniqueGenres = set()
uniqueLanguages = set()
for movie in movies:
    uniqueGenres.update(returnGenre(movie.genres))
    if isinstance(movie.lang, str):
        uniqueLanguages.add(movie.lang)
```

```python
uniqueGenres
```

```python
uniqueGenresDict = {}
for movie in movies:
    if movie.lang == 'en':
        num_female = len(movie.cast.get_female_cast())
        num_male = len(movie.cast.get_male_cast())
        if num_male == 0:
            if num_female > 0:
                ratio = 1.0
            else:
                ratio = 0.0
        else:
            ratio = float(num_female)/num_male
        genres = returnGenre(movie.genres)
        for g in genres:
            if g in uniqueGenresDict:
                uniqueGenresDict[g] += [ratio]
            else:
                uniqueGenresDict[g] = [ratio]
```

```python
for genre in uniqueGenresDict:
    uniqueGenresDict[genre] = round(np.mean(uniqueGenresDict[genre]), 3)
```

# 3. Find female-to-male cast ratio for each language

```python
uniqueLanguagesDict = {}
for movie in movies:
    num_female = len(movie.cast.get_female_cast())
    num_male = len(movie.cast.get_male_cast())
    if num_male == 0:
        if num_female > 0:
            ratio = 1.0
        else:
            ratio = 0.0
    else:
        ratio = float(num_female)/num_male
    lang = movie.lang
    if lang in uniqueLanguagesDict:
        uniqueLanguagesDict[lang] += [ratio]
    else:
        uniqueLanguagesDict[lang] = [ratio]
for lang in uniqueLanguagesDict:
    uniqueLanguagesDict[lang] = round(np.mean(uniqueLanguagesDict[lang]), 3)
```
